```python
from datasets import DatasetDict, load_from_disk, load_metric

datasets = load_from_disk('../data/train_dataset')
# training_dataset= 
training_dataset = datasets['train']
valid_dataset = datasets['validation']
```

```python
li = training_dataset['context'].copy()
print(len(li))
li.extend(valid_dataset['context'])
print(len(li))
```

```python
x= set()
for context in li:
    # print(context)
    a= re.findall(r'[^\w\s]',context)
    x= x.union(set(a))
x
```
